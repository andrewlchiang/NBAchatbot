```python
links = My_table.find_all('a')
links
```

```python
final = {}
# make an iterable version
for team, url in team_url:
    website_url = requests.get(url).text
    soup = BeautifulSoup(website_url, 'html')
    My_table = soup.find('table',{'class':'Table'})
    links = My_table.find_all('a')
    players = []
    for link in links:
        if link.get('class') == ['AnchorLink']:
            players.append(link.get_text('title'))
    players = [i for i in players if i] # removes empty strings lol this code is so PyTHoNIc
    coach = soup.find('p', {'class': "TableDetails__Paragraph"}).get_text()
    players.append(re.sub("Coach: ", "", coach))
    final.update({team:players}) # updates final dictionary
```
